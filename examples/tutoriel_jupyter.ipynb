## 🚀 Partie 1 — Traçage de fonctions

La fonctionnalité la plus simple : observer quand une fonction est appelée.

````python
import sys, os
sys.path.insert(0, os.path.join(os.getcwd(), '..'))

from attovisio import Tracer, reset
reset()  # Nettoie les événements précédents

# Le décorateur @Tracer.trace est tout ce qu'il faut !
@Tracer.trace
def calculer_surface(longueur, largeur):
    """Calcule la surface d'un rectangle."""
    return longueur * largeur

@Tracer.trace
def calculer_périmètre(longueur, largeur):
    """Calcule le périmètre d'un rectangle."""
    return 2 * (longueur + largeur)

# Appelons ces fonctions
s = calculer_surface(5, 3)
p = calculer_périmètre(5, 3)

print(f'Surface   = {s}')
print(f'Périmètre = {p}')
````

## 📖 Partie 2 — Voir les événements capturés

````python
from attovisio import EventBus

bus = EventBus.get_global()
events = bus.get_events()

print(f'Nombre total d\'événements : {len(events)}\n')
for e in events:
    print(e)
````

## 🗣️ Partie 3 — Narration en français

````python
from attovisio import Narrator

narrator = Narrator()
story = narrator.tell()
print(story)
````

## 💥 Partie 4 — Capturer les erreurs

````python
reset()

tracer = Tracer(verbose=True)

@tracer.watch
def diviser(a, b):
    return a / b

# Division normale
diviser(10, 2)

# Division par zéro → l'erreur est capturée ET remontée
try:
    diviser(10, 0)
except ZeroDivisionError:
    print('\n(L\'erreur a bien été capturée par AttoVisio ☝️)')
````

## 📊 Partie 5 — Visualisation HTML interactive

````python
from attovisio import Visualizer
from IPython.display import IFrame, display

viz = Visualizer()

# Générer la timeline
viz.timeline('timeline.html', title='Mon Programme')

# Afficher dans le notebook
display(IFrame('timeline.html', width='100%', height=500))
print('\n✅ timeline.html créé dans le répertoire courant')
````

````python
# Tableau de bord complet
viz.dashboard('dashboard.html', title='Tableau de Bord')
display(IFrame('dashboard.html', width='100%', height=600))
````

## 📂 Partie 6 — Observer les fichiers

````python
reset()
from attovisio import IOObserver

observer = IOObserver(verbose=True)

# Journaliser manuellement des opérations
observer.log_open('rapport.csv', mode='r')
observer.log_read('rapport.csv', size_bytes=1024)
observer.log_write('résultat.txt', size_bytes=512)
observer.log_delete('tmp_cache.pkl')

print('\nRécit des opérations fichiers :')
narrator = Narrator()
file_events = EventBus.get_global().get_events()
for e in file_events:
    print(' ', narrator.narrate(e))
````

## 🔐 Partie 7 — Audit de sécurité

````python
reset()
from attovisio import SecurityAuditor
from attovisio.events import Event

bus = EventBus.get_global()
auditor = SecurityAuditor(bus=bus, verbose=True)
auditor.start()

# Simuler un accès à un fichier sensible
bus.emit(Event(
    kind='file_open',
    data={
        'path': '/home/user/.aws/credentials',
        'absolute_path': '/home/user/.aws/credentials',
        'filename': 'credentials',
        'extension': '',
        'mode': 'r'
    }
))

auditor.stop()
print('\n' + auditor.report())
````

## 💾 Partie 8 — Export JSON

````python
import json

bus = EventBus.get_global()
bus.export_json('events.json')

with open('events.json') as f:
    data = json.load(f)

print(f'{len(data)} événements exportés dans events.json')
print('\nPremier événement :')
print(json.dumps(data[0], indent=2, ensure_ascii=False))
````

## ⚡ Partie 9 — Surveillance système (nécessite psutil)

````python
try:
    from attovisio import Monitor
    import time

    reset()
    
    with Monitor(interval=0.1) as m:
        # Simuler un traitement
        total = sum(i**2 for i in range(100_000))
        time.sleep(0.3)

    summary = m.summary()
    print('📊 Résumé des ressources :')
    print(f"  CPU    moy : {summary['cpu_percent']['avg']}%")
    print(f"  CPU    max : {summary['cpu_percent']['max']}%")
    print(f"  RAM    max : {summary['memory_rss_mb']['max']} MB")
    print(f"  Durée      : {summary['duration_s']} s")

except ImportError:
    print('psutil non installé. Exécutez : pip install psutil')
````

## 🎓 Récapitulatif

| Ce que vous voulez faire | Code AttoVisio |
|--------------------------|----------------|
| Tracer une fonction | `@Tracer.trace` |
| Voir les événements | `EventBus.get_global().get_events()` |
| Narrer en français | `Narrator().tell()` |
| Générer une timeline HTML | `Visualizer().timeline('out.html')` |
| Observer les fichiers | `IOObserver(verbose=True)` |
| Audit de sécurité | `SecurityAuditor(verbose=True).start()` |
| Surveiller le CPU/RAM | `with Monitor() as m: ...` |
| Exporter en JSON | `EventBus.get_global().export_json('f.json')` |
| Réinitialiser | `reset()` |
